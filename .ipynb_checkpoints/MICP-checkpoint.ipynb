### To check for outliers in charges

```python
sb.boxplot(data['charges'])
```

```python
# Finding outliers through IQR Method
```

```python
data_chr = data['charges']
type(data_chr)
```

```python
out_df = pd.DataFrame()
out_df['Columns']= ['charges']
out_df.set_index('Columns', inplace = True)
out_df['Min'] = data_chr.min()
out_df['Max']= data_chr.max()
out_df['Range']= data_chr.max()-data.min()
out_df['Q1'] = data_chr.quantile(.25)
out_df['Q3'] = data_chr.quantile(.75)
out_df['IQR'] = data_chr.quantile(.75) - data_chr.quantile(.25)
out_df['LF'] = out_df['Q1']-(1.5 * out_df['IQR'])
out_df['UF'] = out_df['Q3']+(1.5 * out_df['IQR'])
out_df
```

```python
def outlier(x,lf,uf):
    LO,UO = [],[]
    for i in data[x]:
        if i < lf:
            LO.append(i)
        elif i>uf:
                UO.append(i)
               
                return LO,UO
```

```python
outlier_list = []
outlier_list.append( outlier('charges',-13147.17675,34412.75325))
out_df['outlier'] = outlier_list
out_df
```

## PREDICTION USING LINEAR REGRESSION MODEL

```python
x = data.drop(columns = 'charges', axis = 1)
x
```

```python
y = data['charges']
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,random_state = 1)
x_train.shape
```

```python
y_train.shape
```

```python
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
linreg.fit(x_train,y_train)
```

```python
linreg.intercept_
```

```python
linreg.coef_
```

```python
y_pred = linreg.predict(x_test)
y_pred
```

```python
plt.scatter(y_test, y_pred, color='green')
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')
plt.xlabel('Actual Charges Values')
plt.ylabel('Predicted Charges Values')
```

### MODEL EVALUATION

#### Through MAE (Mean Absolute Error)

```python
from sklearn import metrics
print (metrics.mean_absolute_error(y_test, y_pred))
```

#### Through MSE (Mean Squared Error)

```python
MSE = metrics.mean_squared_error(y_test, y_pred)
MSE
```

#### Through RMSE (Root Mean Squared Error)

```python
RMSE = np.sqrt(MSE)
RMSE
```

#### Through R2 (Goodness of fit test)

```python
R2 =metrics.r2_score(y_test, y_pred)
R2
```

```python
# Conclusion : R2 value is 0.75 which is a good value and the model fits into the data. (The value of R2 should be close to one for better model)
```
